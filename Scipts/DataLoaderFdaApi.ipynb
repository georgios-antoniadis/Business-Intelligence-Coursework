3. Get the zip file, extract it and save it to the designated directory

```python
#from zipfile import ZipFile
# Create a ZipFile Object and load sample.zip in it
#with ZipFile(url, 'r') as zipObj:
   # Extract all the contents of zip file in current directory
#   zipObj.extractall()

#https://stackoverflow.com/questions/5710867/downloading-and-unzipping-a-zip-file-without-writing-to-disk

####################################################################### 
#######################################################################    
def getAndExtract(dataframe):
    
    r = requests.get(dataframe) #links_df['file'][120]
    z = zipfile.ZipFile(io.BytesIO(r.content))
    z.extractall()

####################################################################### 
####################################################################### 
```

4. Open the file and load it to a pandas dataframe

```python
# Opening JSON file
#ONE WAY
#json.load(open('animalandveterinary-event-0001-of-0001.json'))

#2ND WAY

def openAndLoad(jsonfile):
    
    f = open(jsonfile)#'animalandveterinary-event-0001-of-0001.json')
    
    # returns JSON object as
    # a dictionary
    data = json.load(f)
    data['results']
    return pd.json_normalize(data['results'])

#df.iloc[0:4]
```

5. Create some empty dataframes

```python
reaction_df = pd.DataFrame()
outcome_df = pd.DataFrame()
drug_df = pd.DataFrame()
active_ingredients_df = pd.DataFrame()
```

```python
for linksIndex, row in links_df.iterrows():
    print(row['file'])
    print(linksIndex)
    getAndExtract(links_df.iloc[linksIndex]['file'])
```

```python
for linksIndex, LinksRow in links_df.iterrows():
    print(LinksRow['file'])
    print(linksIndex, LinksRow)

    getAndExtract(links_df.iloc[linksIndex]['file'])
    periodValue = links_df['display_name'][linksIndex][0:4]+links_df['display_name'][linksIndex][5:7]

    df = openAndLoad('animalandveterinary-event-0001-of-0001.json')

    for index, row in df.iterrows():
        #print(index, row)
        #assign the unique aer number to a variable
        unique_aer_id_number = df['unique_aer_id_number'][index]

        #normalize the current line's data and load them in the temp dataframe
        reaction_df_temp = pd.json_normalize(df.loc[index,'reaction'])

        #append the column with the unique aer number
        reaction_df_temp['p_recordID'] = unique_aer_id_number
        reaction_df_temp['Period'] = periodValue

        #Append rows to the final dataframe
        reaction_df = reaction_df.append(reaction_df_temp)


        if type(df.loc[index,'outcome']) != float:
            outcome_df_temp = pd.json_normalize(df.loc[index,'outcome'])
            outcome_df_temp['p_recordID'] = unique_aer_id_number
            outcome_df_temp['Period'] = periodValue
            outcome_df = outcome_df.append(outcome_df_temp)

        #if type(df.loc[index,'drug']) != float:
        #    drug_df_temp = pd.json_normalize(df.loc[index,'drug'])
        #    drug_df_temp['p_recordID'] = unique_aer_id_number
        #    drug_df_temp['Period'] = links_df['display_name'][0][0:4]+links_df['display_name'][0][5:7]
        #    drug_df = drug_df.append(drug_df_temp)

        if type(df.loc[index, 'drug']) != float:
            drug_df_temp = pd.json_normalize(df.loc[index,'drug'])
            drug_df_temp['p_recordID'] = unique_aer_id_number
            drug_df_temp['Period'] = periodValue
            drug_df = drug_df.append(drug_df_temp)

            active_ingredients_df_temp = pd.json_normalize(df.loc[index,'drug'],record_path=['active_ingredients'])
            active_ingredients_df_temp['p_recordID'] = unique_aer_id_number
            active_ingredients_df_temp['Period'] = periodValue
            active_ingredients_df = active_ingredients_df.append(active_ingredients_df_temp)
```
